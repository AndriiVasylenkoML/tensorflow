````python
import numpy as np
import keras
````

Первая строка определяет саму модель. Модель — это обучаемая нейронная сеть, в данном случае простейшая нейронная сеть, которая только бывает.
`model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])`

Эта сеть стоить всего из одного условия. Это задается вызовом `keras.layers.Dense(units=1, input_shape=[1])` В єтом условие находится всего 1 нейрон это задано параметром `units=1`
Также мы подаем единственное значение в нейронную сеть значение переменной x.
И мы предлагаем нашей нейронной сети предсказать, чему равно y для данного x.
Поэтому ми задаем значение параметра `input_shape=[1]`.

`model.compile(optimizer='sgd', loss='mean_squared_error')` Для подготовки модели нам необходимо передать 2 функции: Функции потерь и оптимизации. Это ключевее функции в машинном обучении, суть которого состоит в том, что модель пытается угадать зависимость между переменными.

Например, модель может предположить что зависимость y = 5x + 5. В процессе обучения функция потерь оценивает, насколько это хорошее или плохое предположение.
Затем функция оптимизации создаст для нас следующее предположение.
Логика в том, что, работая поочередно функции потерь и оптимизации приведут нас к правильной формуле.

````python
model = keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')
````


В нашем случае система повторит это цикл 500 раз. Она создает начальное предположение, оценивает его качество,  используя функции потерь, улучшает его,  используя оптимизатор и так далее.`model.fit(xs, ys, epochs=500)`  

Данные задаются 2-мя массивами данными xs  и ys. Код, который их сопоставляет, находится внутри метода fit нашей модели. В буквальном смысле мы подгоняем наши x и y друг к другу 500 раз.

``` python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=500)
```

````python
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

model.fit(xs, ys, epochs=500)
````

Когда процесс завершен, у нас есть обученная модель. Теперь ми можем использовать эту модель для предсказания значения y для заданного х. Как вы при выполнении этой строчки кода, где мы говорим: "Предскажи, значение y, если значение x = 10". `print(model.predict([10.0]))`

Близко, но не точно. Почему, как вам кажется, это происходит?
Компьютер был обучен всего на 6 паре чисел. Похоже, что все эти 6 точек лежат на прямой лини. Но возможно, что другие примеры не буду лежать на той же линии. Существует высокая вероятность того эта прямая линия, но мы не можем знать наверняка. И это заложено в нашем предсказании. Полученное значение близко к 19,  но не равно точно этому знанчению.
